```python
def fractional_knapsack(values, weights, capacity):
    ratio = [(values[i] / weights[i], values[i], weights[i]) for i in range(len(values))]
    ratio.sort(reverse=True)

    total_value = 0
    remaining = capacity
    for r, v, w in ratio:
        if remaining == 0:
            break
        if w <= remaining:
            total_value += v
            remaining -= w
            print(f"Took full item (value={v}, weight={w}) → Remaining capacity={remaining}")
        else:
            fraction = remaining / w
            total_value += v * fraction
            print(f"Took {fraction:.2f} fraction of item (value={v}, weight={w})")
            remaining = 0
    print(f"\nMaximum value in knapsack ((0.67*120)+60+100) = {total_value:.2f}") #remove this formula

values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50

fractional_knapsack(values, weights, capacity)
```
