```python
def fib_recursive(n, steps):
    steps[0] += 1
    if n <= 1:
        return n
    return fib_recursive(n - 1, steps) + fib_recursive(n - 2, steps)

def fib_iterative(n):
    steps = 0
    a, b = 0, 1
    fib_series = [0]
    if n >= 1:
        fib_series.append(1)

    print("\nIterative Calculation:")
    print("F(0) = 0")
    if n > 0:
        print("F(1) = 1")

    for i in range(2, n + 1):
        steps += 1
        c = a + b
        print(f"Step {steps}: {a} + {b} = {c}")
        fib_series.append(c)
        a, b = b, c

    return fib_series, steps


# Input from user
n = int(input("Enter the value of n: "))



# Iterative calculation
fib_series, steps_itr = fib_iterative(n)

# Print full series
print("\nFull Fibonacci Series:")
for i, val in enumerate(fib_series):
    print(f"F({i}) = {val}")

print(f"Step count = {steps_itr}")              #remove the formula
```
