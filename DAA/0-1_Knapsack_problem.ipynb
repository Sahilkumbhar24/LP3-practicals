```python
def knapsack(values, weights, capacity):
    n = len(values)
    K = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                K[i][w] = max(values[i-1] + K[i-1][w - weights[i-1]], K[i-1][w])
            else:
                K[i][w] = K[i-1][w]

    print("DP Table:")
    for row in K:
        print(row)
    print("\nSo we take items 2 and 3 (weights 2 + 3 = 50 → total value 220(100+120)")     #Remove this
    print(f"\nMaximum value in knapsack = {K[n][capacity]}")
    return K[n][capacity]

values = [60, 100, 120]
weights = [1, 2, 3]
capacity = 5

knapsack(values, weights, capacity)
```
